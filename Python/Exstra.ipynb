```python
# xmlデータからモジュールの説明を抜き出す。
import pandas as pd
import xmljson
from lxml import etree
import os

def GetModuleDescriptionStrings(targetFiles):
  outputData = [["ModuleName", "JPN", "ENG"]]
  for xml in targetFiles:
    xml_tree = etree.parse(xml)
    xml_root = xml_tree.getroot()
    xml_dict = xmljson.yahoo.data(xml_root)

    descTextMod = xml_dict['EDSInfo']['AdapterInfo']['DescTextMulLan']
    productName = xml_dict['EDSInfo']['AdapterInfo']['ProdName']

    # print(descTextMod)
    # print(productName)

    data = [productName]
    for i, lang in enumerate(descTextMod):
      data.append(descTextMod[lang])
    outputData.append(data)
    # print(outputData)
  # print(outputData)
  df = pd.DataFrame(outputData)
  print(df)
```

```python
import pandas as pd
import xmljson
from lxml import etree
import os

def GetModuleParamInfoStrings(targetFiles):
  outputData = [["", "Name(JPN)", "Name(ENG)", "Help(JPN)", "Help(ENG)", "Enum(JPN)", "Enum(ENG)" ]]
  for xml in targetFiles:
    xml_tree = etree.parse(xml)
    xml_root = xml_tree.getroot()
    xml_dict = xmljson.yahoo.data(xml_root)

    paramInfos = xml_dict['EDSInfo']['ParameterTable']['Parameter']
    productName = xml_dict['EDSInfo']['AdapterInfo']['ProdName']

    # print(paramInfos)
    # print(productName)

    for i, param in enumerate(paramInfos):
      data = [productName]
      paramName = param['ParameterNameMulLan']
      for lan in paramName:
        data.append(paramName[lan])

      paramHelp = param['HelpStringMulLan']
      for lan in paramHelp:
        data.append(paramHelp[lan])

      enumCandidate = param['EnumMulLan']
      for lan in enumCandidate:
        data.append(enumCandidate[lan])
      # print(data)
      outputData.append(data)
      #data.clear()
    # 
    # print(outputData)
  df = pd.DataFrame(outputData)
  
  # desktop = os.path.join(os.path.join(os.environ['USERPROFILE']), 'Desktop')
  df.to_csv('C:\\Users\\aksh0\\OneDrive\\デスクトップ\\output.csv', encoding="shift-jis")
  print(df)
```
